```python
"""Q1. (Data Loading & Initial Analysis)
Load the Ford Car Dataset using pandas.
Display the first 10 rows and last 5 rows.
Check the shape (number of rows and columns) and data types of all
columns.
Write observations about the dataset structure in comments.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from google.colab import files

uploaded = files.upload()

for fn in uploaded:
  print(f"Uploaded file:{fn}")
filename = list(uploaded)[0]

data = pd.read_csv(filename)
print(f"First 10 rows:\n{data.head(10)}")
print(f"Last 5 rows:\n{data.tail(5)}")
print(f"Shape of the data:{data.shape}")
print(f"Data  types of all columns:\n{data.dtypes}")

# after observin the output the dataset contain information about ford car collection
# in the dataset there are 9 columns and 17966 rows

```

```python
"""Q2. (Missing & Duplicate Values)
Perform a detailed check on the dataset:
Find the total number of missing values in each column.
Identify and remove duplicate rows if any.
Write a summary of your findings and how you handled them in comments.
 """

print(f"Total number of the missing values in each column:\n{data.isnull().sum()}")
print("Identifying and remove duplicate rows if any.")
if data.duplicated().sum():
  print(f"{data.duplicated().sum()} duplicate rows found")
  data.drop_duplicates(inplace=True)
  print("Duplicate rows are removed")
  print(f"After removind:{data.duplicated().sum()} are duplicated")

else:
  print("no duplicate rows are found.")
  print(f"Count of duplicate:{data.duplicated().sum()}")

# int the dataset I find 154 duplicate rows and 0 null values all duplicate rows using drop_duplicate() method
```

```python
"""Q3. (Statistical Summary)
Generate the statistical summary of all numeric columns using describe().
Identify minimum, maximum, mean, and median values for key columns like price, mileage, and year.
"""

print(f"Statistical summary of all numeric columns:\n{data.describe()}")
print("*"*30)
columns = ["price","mileage","year"]
for column in columns:
  print("*"*30)
  print(f"Minimum value of the :{column} is:{data[column].min()}")
  print(f"Maximum value of the :{column} is:{data[column].max()}")
  print(f"Mean value of the :{column} is:{data[column].mean()}")
  print(f"Median value of the :{column} is:{data[column].median()}")
  print("*"*30)


```

```python
"""Q4. (Histogram of Numeric Features)
Plot histograms for all important numeric columns (price, mileage, year,
engineSize, mpg).
Analyze the distribution for each feature and note your findings in comments.
"""

column = ['price','mileage','year','engineSize','mpg']
plt.figure(figsize=(10,8))
for fn in column:
  plt.subplot(2,3,column.index(fn)+1)
  plt.title(f"Histogram of {fn}")
  sb.histplot(data[fn],kde=True)
plt.tight_layout()
plt.show()
# in the graps we find that most of the pruchase their car between 1500 to 2000
# in the histogram of mileage the count of the velicals having mileage 1 to 3000 is very heigh
# in the histograph of the year cars buy after 2026 are false data which need to be remove
# in the histograph of the engine size data of the enginesize = 0 need to correct
# in the histograph of the mpg data of mpg = 0 also need to correct
```

```python
"""Q5. (Count Plots of Categorical Features)
Create count plots for all categorical columns (fuelType, transmission, model, etc.) using seaborn.
Analyze which categories are most common.
Write insights about market trends based on these plots in comments.
"""
catagorical_cols = data.select_dtypes(include='object').columns
plt.figure(figsize=(10,10))
for fn in catagorical_cols:
  plt.subplot(2,2,catagorical_cols.tolist().index(fn)+1)
  sb.countplot(x=fn,data=data)
  plt.title(f"Countplot of :{fn}")
  plt.xlabel(f"{fn}")
  plt.ylabel("Count")
plt.tight_layout()
plt.show()

# in contplot of the model there are 23 different modals which makes chart unreadabl praphs fiesta is sold most than any other model
# in countplot of the transmission manual transmission has highest demand
# in the count plot of fule type petrol cars mostly used followed by diesel cars
```

```python
"""Q6. (Correlation Heatmap)
Create a heatmap of the correlation matrix for numeric features.
Use seaborn with proper annotations.
Identify which features are highly correlated with the target variable (price). Write your observations in comment.
"""

corelation_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sb.heatmap(corelation_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
# The tax,mileage and the year is highely corealated ith the variable price here year and tax show
# positive corelation and mileage show negative corealtion
#  in real world year only show positive corealation with price in case on special cars like 'konisec'
# but useally with most of the cars as years grow up car price reduce
```

```python
"""Q8. (Encoding Categorical Variables)
Convert all categorical columns into numeric form using appropriate encoding
techniques:
Use One Hot Encoding.
Show before and after transformation for at least 2 columns."""

print(f"Transmission , fule type of the car before encoding:\n{data[['transmission','fuelType']].sample(10)}")
print("*"*30)
data_encoded = pd.get_dummies(data)
print(f"Categorical columns  of the car after encoding:\n{data_encoded.sample(10)}")
```

```python
"""Q9. (Feature Scaling)
Apply Standard Scaling (using StandardScaler from sklearn) on the numeric
independent features.
Show the first 5 rows of scaled data."""

# converging all boolean values into int
data_encoded.astype(int)
# applying scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_encoded[['price','mileage','year','tax','mpg','engineSize']] = scaler.fit_transform(data_encoded[['price','mileage','year','tax','mpg','engineSize']])
print(f"After scaling first 5 rows of the data:\n{data_encoded.head()}")
```
